```python
import pickle
import networkx as nx
import operator
from networkx.readwrite import json_graph
import json
```

```python
#Leemos el grafo
G=nx.read_gpickle("GrafoRelaciones.pkl")
```

```python
print(nx.info(G))
```

```python
nx.density(G)
```

## Medidas de centralidad

### Degree centrality

```python
centr=nx.degree_centrality(G)
```

```python
sorted_centr = sorted(centr.items(), key=operator.itemgetter(1), reverse=True)
sorted_centr
```

### Closeness

```python
closen=nx.closeness_centrality(G)
```

```python
sorted(closen.items(), key=operator.itemgetter(1), reverse=True)
```

### Betweenness

```python
bet=nx.betweenness_centrality(G) 
```

```python
sorted(bet.items(), key=operator.itemgetter(1), reverse=True)
```

### Eigenvector centrality

```python
eigencent=nx.eigenvector_centrality(G)
```

```python
sorted(eigencent.items(), key=operator.itemgetter(1), reverse=True)
```

## Componentes conexas

```python
#Número de componentes conexas
nx.number_connected_components(G)
```

```python
cc=[G.subgraph(c) for c in nx.connected_components(G)]
```

```python
cc[1].nodes
```

```python
len(cc[1].edges)
```

```python
[c.nodes for c in cc if len(c)>10 and len(c)<100]
```

```python
#Tamaño de las componentes conexas
[len(comp) for comp in cc]
```

## Escribir subgrafo

```python
def escribe_grafo(g, fichero):
    data1 = json_graph.node_link_data(g)
    with open(fichero, "w") as f:
        f.write(json.dumps(data1))
```

```python
escribe_grafo(cc[1], "componente11.json")
```
